```python
import zipfile
import os

zip_file_path = '/content/drive/MyDrive/Smart Football- Object Detection.v11-od_sgd_optimizer.yolov8.zip'
extraction_path = '/content/'

# Create the extraction directory if it doesn't exist
os.makedirs(extraction_path, exist_ok=True)

with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extraction_path)

print(f"'{zip_file_path}' has been extracted to '{extraction_path}'.")
print(f"Contents of '{extraction_path}':")
!ls -F {extraction_path}
```

```python
from ultralytics import YOLO

# load model pretrained
model = YOLO("yolov8n.pt")  # nhẹ, nhanh

# train
model.train(
    data="/content/data.yaml",
    epochs=50,
    imgsz=640,
    batch=8
)
```
